```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("student-mat.csv", sep=';')


print("Shape of dataset:", df.shape)

print("\nColumns in dataset:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nMissing values:")
print(df.isnull().sum())

avg_grade = df['G3'].mean()
print("\nAverage Final Grade:", avg_grade)


above_15 = df[df['G3'] > 15].shape[0]
print("Students scoring above 15:", above_15)


correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and grade:", correlation)

gender_performance = df.groupby('sex')['G3'].mean()
print("\nGender-wise performance:")
print(gender_performance)

plt.figure()
plt.hist(df['G3'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()

plt.figure()
gender_performance.plot(kind='bar')
plt.title("Average Grades by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

## Conclusion

- The dataset contains student performance data.
- The average final grade is around 10.4.
- Some students scored above 15, indicating high performance.
- There is a positive correlation between study time and grades.
- Gender-wise performance shows slight variation between male and female students.
